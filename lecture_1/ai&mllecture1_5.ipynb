```python
#q6. body mass index(bmi) is a simple measure that helps determine  whether a
# person's weight is appropriate for their height .
# write a pyhon program that takes the bmi value as input from the user
# and displays the corresponding weight category using the following rules :
# bmi value category
# less than 18.5 underweight
# 18.5 to 24.9 normal weight
# 25.0 to 29.9 overweight
# 30.0 or above obese
```

```python
height=float(input("enter the height(m) "))
weight=float(input("enter the weight in(kg) "))
bmi=weight/(height*height)
if(bmi<18.5):
  print("underweight",bmi)
elif(bmi<=24.9):
  print("normal weight ",bmi)
elif(bmi<=29.9):
  print("overweight ",bmi)
else:
  print(" obese",bmi)

```
